```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the cleaned dataset
zip_file_path = r"C:\Users\tanzi\Personal Projects\Market-Basket-Analysis\data\cleaned_transactions.zip"

# Extract and read CSV from ZIP
import zipfile
with zipfile.ZipFile(zip_file_path, 'r') as z:
    with z.open('cleaned_transactions.csv') as f:
        df = pd.read_csv(f, parse_dates=['date'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Grouping orders by Customer
clv_data = df.groupby("customer_name").agg({
    "date": ["min", "max"], # First & last purchase date
    "total_cost": ["sum", "count"] # Total amount spent & number of transactions
}).reset_index()

# Renaming Columns 
clv_data.columns = ["customer_name", "first_purchase", "last_purchase", "total_spent", "total_orders"]
```

```python
clv_data.head()
```

```python
from datetime import datetime

# Latest date in the dataset
max_date = df["date"].max()

# Calculating CLV Features
clv_data["recency"] = (clv_data["last_purchase"] - clv_data["first_purchase"]).dt.days
clv_data["T"] = (max_date - clv_data["first_purchase"]).dt.days  # Tenure of customer --> (the number of days since the first purchase until the latest transaction in the dataset (max_date).)
clv_data["frequency"] = clv_data["total_orders"] - 1  # Repeat purchases
clv_data["monetary_value"] = clv_data["total_spent"] / clv_data["total_orders"]  # Avg. spend per order
```

```python
clv_data.head()
```

```python
# Filter customers with at least one repeat purchase
clv_data = clv_data[clv_data["frequency"] > 0]

# After filtering
print(f"Total Customers Eligible for CLV Modeling: {len(clv_data)}")
clv_data.head()
```

```python
!pip install lifetimes
# Import CLV models
from lifetimes import BetaGeoFitter, GammaGammaFitter
```

```python
# Training the BG/NBD model
bgf = BetaGeoFitter()
bgf.fit(clv_data["frequency"], clv_data["recency"], clv_data["T"])

# Predicting expected purchases for the next 3 months
clv_data["predicted_purchases_3m"] = bgf.predict(90, clv_data["frequency"], clv_data["recency"], clv_data["T"])
```

```python
# Show top customers by predicted purchases
clv_data.sort_values("predicted_purchases_3m", ascending=False).head(10)
```

```python
# Checking for customers with zero or negative monetary value and removing them
print("Customers with zero or negative monetary value:", (clv_data["monetary_value"] <= 0).sum())
clv_data = clv_data[clv_data["monetary_value"] > 0]
```

```python
print("Remaining customers:", len(clv_data))
clv_data.head()
```

```python
# Train Gamma-Gamma Model with a penalizer to improve convergence -->(did not converge previously)
ggf = GammaGammaFitter(penalizer_coef=0.01)  # Adding a small penalty

# Fitting the model
ggf.fit(clv_data["frequency"], clv_data["monetary_value"])

# Predicting Average Spending per Customer with Gamma-Gamma Model
clv_data["predicted_monetary_value"] = ggf.conditional_expected_average_profit(clv_data["frequency"], clv_data["monetary_value"])
```

```python
clv_data.sort_values("predicted_monetary_value", ascending=False).head(10)
```

```python
# Calculating CLV
clv_data["CLV"] = clv_data["predicted_purchases_3m"] * clv_data["predicted_monetary_value"]

# Plotting CLV distribution
plt.figure(figsize=(10, 5))
sns.histplot(clv_data["CLV"], bins=50, kde=True)
plt.title("Distribution of Customer Lifetime Value (CLV)")
plt.xlabel("Customer Lifetime Value ($)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Show top 10 customers by CLV
clv_data.sort_values("CLV", ascending=False).head(10)
```

```python
# Means a small number of customers have very high CLV
```

```python
# Define CLV percentiles
high_threshold = clv_data["CLV"].quantile(0.80)  # Top 20%
low_threshold = clv_data["CLV"].quantile(0.40)   # Bottom 40%

# Categorize customers
def categorize_clv(clv):
    if clv >= high_threshold:
        return "High CLV (VIP)"
    elif clv < low_threshold:
        return "Low CLV"
    else:
        return "Medium CLV"

clv_data["CLV_Segment"] = clv_data["CLV"].apply(categorize_clv)
```

```python
# Showing segment distribution
print(clv_data["CLV_Segment"].value_counts())
```

```python
clv_data.head(10)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=clv_data, x="CLV_Segment", palette="coolwarm", order=["Low CLV", "Medium CLV", "High CLV (VIP)"])
plt.title("Customer Segments Based on CLV")
plt.xlabel("CLV Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Strategies for each CLV segment
def clv_strategy(segment):
    if segment == "High CLV (VIP)":
        return "Exclusive rewards, premium offers, priority service"
    elif segment == "Medium CLV":
        return "Upsell/Cross-sell promotions, loyalty programs"
    else:
        return "Win-back campaigns, discounts, retention offers"

# Apply strategy
clv_data["Business_Strategy"] = clv_data["CLV_Segment"].apply(clv_strategy)
```
